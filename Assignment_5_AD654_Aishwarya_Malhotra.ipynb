##### Solution - G

```python
# a] Calculate the 15 - period moving average

sny["Close_15_MA"] = sny["Close"].rolling(window=15).mean()

plt.figure(figsize=(10,6))
plt.plot(sny["Close"], label="Acutal Closing Prices", color="orange")
plt.plot(sny["Close_15_MA"], label="15 - Period Moving Average", color="green")
plt.title('15 - Period Moving Average vs Actual Closing Prices')
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.legend()
plt.show()
```

```python
# b] Calculating 65 - period moving average


sny["Close_65_MA"] = sny["Close"].rolling(window=65).mean()

plt.figure(figsize=(10, 6))
plt.plot(sny["Close"], label="Actual CLosing Prices", color="cyan")
plt.plot(sny["Close_65_MA"], label="65 - Period Moving Average", color="magenta")
plt.title("65 - Period Moving Average vs Actual Closing Prices")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.legend()
plt.show()
```

**Solution - c]-**

**Difference between the two moving average plots**

**Smoothing -** The 15 - period moving average is more volatile than price change compared to the 65 - period moving average. This is because plot a] uses a shorter window, incorporating fewer data points. The shorter window makes it more susceptible to short-term price fluctuations. Plot b] with its longer window, averages over a larger timeframe, smoothing out those short-term fluctuations and presenting a more stable line.

**Responsiveness -** Due to the shorter window, plot a] is more responsive to recent price changes. It will capture sharp increases or dips in price quicker than plot b]. Conversely, plot b] with its longer window is slower to react to recent price movements. It reflects a more long-term average price trend.

**Shorter Window (15 - period)**

Pros: More responsive to recent price changes, helping identify short-term trends and potential trading opportunities.

Cons: More volatile and prone to sharp price reversals, making it potentially misleading for long term analysis.

**Longer Window (65 - period)**

Pros: Smoother line, making it easier to identicy long-term trends in the price movement. Less susceptible to short-term noise, providing a clearer picture of the overall price direction.

Cons: Slower to react to recent price changes, potentially missing out on short-term trading opportunities.

##### Solution - D

```python
import statsmodels.api as sm
```

```python
X = df[["Web", "Bus_Stop", "Newspaper"]]
y = df['Sales']
```

```python
X = sm.add_constant(X)
```

```python
model = sm.OLS(y, X).fit()
```

```python
model.summary()
```

**Solution-**

**F -  Statistic and Model Significance**

P-value of the (F-statistic):	8.13e-99 (essentially 0)

This extremely small p-value indicates that the model is statistically significant at a very high level of confidence (likely well beyond the typical 0.05 significance level). It is highly unlikely that the observed relationship between the independent variables (Web, Bus Stop, Newspaper) and the dependent variable (Sales) could be due to random chance. This suggests the model has a good fit to the data and can potentially explain a significant protion of the variation in sales.

**Individual Predictor P-values**

Web		    0.000 (highly significant)
Bus_Stop	0.000 (highly significant)
Newspaper	0.954 (marginally significant)

**Web and Bus_Stop:** Both have very low p-value, indicating a statistically significant relationship with sales. This suggests that changes in web and bus stop advertising spending are likely to have a measurable impact on sales.

**Newspaper:** The p-value for newspaper advertising is higher (0.058), which is closer  to the margin of significance (often taken as 0.05). While it suggests a possible relationship, there's a slightly higher chance it could be due to random chance. 

##### Solution - E

```python
df["Interaction"] = df["Web"] * df["Bus_Stop"]
```

```python
X2 = df[["Web", "Bus_Stop", "Interaction"]]
```

```python
X2 = sm.add_constant(X2)
```

```python
model2 = sm.OLS(y, X2).fit()
```

```python
model2.summary()
```

```python
X3 = df[["Web","Bus_Stop"]]
```

```python
model3 = sm.OLS(y, X3).fit()
```

```python
model3.summary()
```

**Solution-**

i]
**Individual Predictor P-values**

Web		    0.000 (highly significant)
Bus_Stop	0.005 (positively significant)
Interaction	0.000 (highly significant)

ii]
**Interaction Term vs No Interaction: R - Squared Comparison**
model2 (with interaction): R-squared = 0.914
model3 (without interaction): R-squared = 0.903

model2 has a slightly higher R-squared compared to model3. While the difference seems small, it sugggests that the interaction term captures an additional effect on sales beyond the individual effects of web and bus stop advertising. In simpler termsm the impact of web advertising on sales might not be constant but depend on the level od bus stop advertising spending, and vice versa. 

```python
web_spending = 150
bus_stop_spending = 24
```

```python
interaction = web_spending * bus_stop_spending
```

```python
sales_predicted = model2.predict([1, web_spending, bus_stop_spending, interaction])[0]
print("Predicted Sales Outcome:", sales_predicted)
```

**Solution-**

iii]

This interaction effect highlights the importance of a multi-channel marketing strategy, By considering how different advertising channels work together, marketers can optimize their approcah. The interaction effect goes beyond simply adding the individual effects of web and bus stop advertising. It suggests that these channels can reinforce see with a single channel approach.
